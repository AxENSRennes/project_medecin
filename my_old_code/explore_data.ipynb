avec les colonnes de validité : si les deux yeux ne sont pas valides? et les pupilles nan? 

```python
# Correction de la syntaxe
blinks = tobii_cleaned[
    (tobii_cleaned['Eye movement type'] == 'EyesNotFound') &
    (tobii_cleaned['Validity left'] == 'Invalid') & 
    (tobii_cleaned['Validity right'] == 'Invalid')&
    (tobii_cleaned['Pupil diameter left'].isna()) &
    (tobii_cleaned['Pupil diameter right'].isna())
]

print(f"Nombre de lignes avec clignements détectés : {len(blinks)}")
blinks.head()

# Identifier les groupes de clignements consécutifs
# (car un clignement peut durer plusieurs échantillons)
timestamp_col = tobii_cleaned.columns[0]

if len(blinks) > 0:
    blinks_sorted = blinks.sort_values(by=timestamp_col)
    blink_timestamps = blinks_sorted[timestamp_col].values
    
    # Grouper les clignements consécutifs (gap < 50ms = même clignement)
    blink_groups = []
    current_group_start = blink_timestamps[0]
    current_group_end = blink_timestamps[0]
    
    for i in range(1, len(blink_timestamps)):
        time_diff = blink_timestamps[i] - blink_timestamps[i-1]
        if time_diff <= 50:  # Même clignement si gap < 50ms
            current_group_end = blink_timestamps[i]
        else:  # Nouveau clignement
            blink_groups.append({
                'start': current_group_start,
                'end': current_group_end,
                'duration': current_group_end - current_group_start
            })
            current_group_start = blink_timestamps[i]
            current_group_end = blink_timestamps[i]
    
    # Ajouter le dernier groupe
    blink_groups.append({
        'start': current_group_start,
        'end': current_group_end,
        'duration': current_group_end - current_group_start
    })
    
    # Calculer le temps moyen entre clignements (entre les groupes)
    if len(blink_groups) > 1:
        intervals = []
        for i in range(1, len(blink_groups)):
            interval = blink_groups[i]['start'] - blink_groups[i-1]['end']
            intervals.append(interval)
        
        intervals = np.array(intervals)
        # Filtrer les intervalles raisonnables (2-10 secondes)
        reasonable_intervals = intervals[(intervals >= 2000) & (intervals <= 10000)]
        
        if len(reasonable_intervals) > 0:
            mean_interval = np.mean(reasonable_intervals)
            median_interval = np.median(reasonable_intervals)
            print(f"\nNombre de clignements uniques détectés : {len(blink_groups)}")
            print(f"Temps moyen entre clignements : {mean_interval:.2f} ms ({mean_interval/1000:.2f} secondes)")
            print(f"Temps médian entre clignements : {median_interval:.2f} ms ({median_interval/1000:.2f} secondes)")
        else:
            print(f"\n⚠️ Intervalles détectés mais hors plage normale")
            print(f"Intervalles moyens (non filtrés) : {np.mean(intervals):.2f} ms")
    else:
        print(f"\n⚠️ Seulement {len(blink_groups)} clignement(s) détecté(s)")
    
    # Durée moyenne des clignements
    blink_durations = [g['duration'] for g in blink_groups]
    print(f"\nDurée moyenne d'un clignement : {np.mean(blink_durations):.2f} ms")
    print(f"Durée médiane : {np.median(blink_durations):.2f} ms")

# Calculer le pourcentage de lignes de clignement
total_lines = len(tobii_cleaned)
blink_lines = len(blinks)
blink_percentage = (blink_lines / total_lines) * 100

print(f"\n{'='*50}")
print(f"Pourcentage de lignes de clignement : {blink_percentage:.2f}%")
print(f"({blink_lines} lignes sur {total_lines} totales)")
```

```python
# Version simplifiée pour visualisation rapide
def plot_column_timeseries(tobii_cleaned, column_index):
    """
    Visualise rapidement une colonne avec gestion des NaN
    """
    timestamp_col = tobii_cleaned.columns[0]
    data_col = tobii_cleaned.columns[column_index]
    
    # Extraire les données
    timestamps = tobii_cleaned[timestamp_col]
    values = pd.to_numeric(
        tobii_cleaned[data_col].astype(str).str.replace(',', '.'), 
        errors='coerce'
    )
    
    # Créer la figure
    plt.figure(figsize=(15, 5))
    
    # Tracer les données valides
    valid_mask = ~values.isna()
    plt.plot(timestamps[valid_mask], values[valid_mask], 
            'b-', linewidth=0.5, alpha=0.7, label='Données valides')
    
    # Interpoler et tracer en pointillés pour montrer les interpolations
    values_interp = values.interpolate(method='linear')
    nan_mask = values.isna()
    if nan_mask.any():
        plt.plot(timestamps[nan_mask], values_interp[nan_mask], 
                'r--', linewidth=0.3, alpha=0.5, label='Interpolation')
    
    plt.xlabel('Recording timestamp')
    plt.ylabel(data_col)
    plt.title(f'Série temporelle - {data_col}')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Afficher les stats
    nan_pct = (values.isna().sum() / len(values)) * 100
    plt.text(0.02, 0.98, 
            f'NaN: {values.isna().sum()}/{len(values)} ({nan_pct:.2f}%)',
            transform=plt.gca().transAxes, 
            verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.show()

# Utilisation simple
plot_column_timeseries(tobii_cleaned, column_index=30) 
plot_column_timeseries(tobii_cleaned, column_index=25)
```

## Interpolation, choix des seuils

```python
def analyze_timestamp_intervals(tobii_data, timestamp_col='Recording timestamp'):
    """
    Analyse les intervalles temporels et recommande un seuil de gap
    """
    timestamps = tobii_data[timestamp_col].values
    intervals = np.diff(timestamps)
    intervals = intervals[intervals > 0]  # Enlever les zéros
    
    print("=" * 60)
    print("ANALYSE DES INTERVALLES TEMPORELS")
    print("=" * 60)
    print(f"\nStatistiques descriptives:")
    print(f"  - Nombre d'intervalles : {len(intervals)}")
    print(f"  - Minimum : {intervals.min():.2f} ms ({intervals.min()/1000:.2f} s)")
    print(f"  - 25ème percentile : {np.percentile(intervals, 25):.2f} ms")
    print(f"  - Médiane : {np.median(intervals):.2f} ms ({np.median(intervals)/1000:.2f} s)")
    print(f"  - 75ème percentile : {np.percentile(intervals, 75):.2f} ms")
    print(f"  - 90ème percentile : {np.percentile(intervals, 90):.2f} ms")
    print(f"  - Maximum : {intervals.max():.2f} ms ({intervals.max()/1000:.2f} s)")
    print(f"  - Moyenne : {intervals.mean():.2f} ms ({intervals.mean()/1000:.2f} s)")
    print(f"  - Écart-type : {intervals.std():.2f} ms")
    
    # Recommandations
    print(f"\n{'='*60}")
    print("RECOMMANDATIONS DE SEUIL:")
    print(f"{'='*60}")
    
    # Option 1 : Médiane
    median_threshold = np.median(intervals)
    gaps_below_median = (intervals <= median_threshold).sum()
    print(f"\n1. Seuil = Médiane ({median_threshold:.2f} ms):")
    print(f"   - {gaps_below_median}/{len(intervals)} gaps seraient interpolés ({gaps_below_median/len(intervals)*100:.1f}%)")
    
    # Option 2 : 75ème percentile
    p75_threshold = np.percentile(intervals, 75)
    gaps_below_p75 = (intervals <= p75_threshold).sum()
    print(f"\n2. Seuil = 75ème percentile ({p75_threshold:.2f} ms):")
    print(f"   - {gaps_below_p75}/{len(intervals)} gaps seraient interpolés ({gaps_below_p75/len(intervals)*100:.1f}%)")
    
    # Option 3 : 90ème percentile
    p90_threshold = np.percentile(intervals, 90)
    gaps_below_p90 = (intervals <= p90_threshold).sum()
    print(f"\n3. Seuil = 90ème percentile ({p90_threshold:.2f} ms):")
    print(f"   - {gaps_below_p90}/{len(intervals)} gaps seraient interpolés ({gaps_below_p90/len(intervals)*100:.1f}%)")
    
    # Option 4 : Pas de limite
    print(f"\n4. Pas de limite (interpoler tous les gaps):")
    print(f"   - Tous les {len(intervals)} gaps seraient interpolés")
    print(f"   - ⚠️ Attention : interpolation sur gaps très longs peut introduire des artefacts")
    
    # Visualisation
    plt.figure(figsize=(12, 5))
    plt.hist(intervals, bins=50, edgecolor='black', alpha=0.7)
    plt.axvline(median_threshold, color='green', linestyle='--', label=f'Médiane ({median_threshold:.0f} ms)')
    plt.axvline(p75_threshold, color='orange', linestyle='--', label=f'75ème percentile ({p75_threshold:.0f} ms)')
    plt.axvline(p90_threshold, color='red', linestyle='--', label=f'90ème percentile ({p90_threshold:.0f} ms)')
    plt.xlabel('Intervalle temporel (ms)')
    plt.ylabel('Fréquence')
    plt.title('Distribution des intervalles temporels')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
    
    return {
        'median': median_threshold,
        'p75': p75_threshold,
        'p90': p90_threshold,
        'intervals': intervals
    }

# Utiliser la fonction
analysis = analyze_timestamp_intervals(tobii_cleaned)
```

```python
def analyze_data_and_suggest_config(tobii_data, timestamp_col='Recording timestamp'):
    """
    Analyse les données et suggère automatiquement les paramètres optimaux
    """
    timestamps = tobii_data[timestamp_col].values
    intervals = np.diff(timestamps)
    intervals = intervals[intervals > 0]
    
    # 1. Détecter l'intervalle de rééchantillonnage optimal
    Q1 = np.percentile(intervals, 25)
    Q3 = np.percentile(intervals, 75)
    IQR = Q3 - Q1
    normal_intervals = intervals[
        (intervals >= Q1 - 1.5*IQR) & 
        (intervals <= Q3 + 1.5*IQR)
    ]
    
    target_interval_ms = np.median(normal_intervals) if len(normal_intervals) > 0 else np.median(intervals)
    sampling_rate_hz = 1000 / target_interval_ms
    
    # 2. Détecter le seuil de gap adaptatif
    max_gap_ms = np.percentile(intervals, 90)  # 90ème percentile
    
    # 3. Suggérer la taille de fenêtre pour smoothing
    # Règle : ~100-200ms de fenêtre pour Savitzky-Golay
    window_duration_ms = 150  # 150ms de fenêtre
    window_size = int(window_duration_ms / target_interval_ms)
    if window_size % 2 == 0:
        window_size += 1  # Doit être impair pour SG
    window_size = max(3, min(window_size, 11))  # Entre 3 et 11
    
    # 4. Durée de baseline (30s ou 10% de la durée totale)
    total_duration_s = (timestamps.max() - timestamps.min()) / 1000
    baseline_duration = min(30, total_duration_s * 0.1)  # Max 30s ou 10% de la durée
    
    # 5. Vérifier si échantillonnage régulier
    cv = np.std(intervals) / np.mean(intervals) if np.mean(intervals) > 0 else 0
    is_regular = cv < 0.1
    
    config = {
        'resampling': {
            'enabled': True,
            'target_interval_ms': target_interval_ms,
            'force': not is_regular
        },
        'missing_data': {
            'max_gap_ms': max_gap_ms,
            'method': 'interpolate'
        },
        'smoothing': {
            'window_size': window_size,
            'poly_order': 2
        },
        'baseline_correction': {
            'baseline_duration': baseline_duration
        },
        'artifact_detection': {
            'z_threshold': 3,  # Standard
            'iqr_factor': 1.5  # Standard
        }
    }
    
    print("=" * 60)
    print("ANALYSE PRÉALABLE ET CONFIGURATION SUGGÉRÉE")
    print("=" * 60)
    print(f"\nCaractéristiques détectées:")
    print(f"  - Échantillonnage régulier : {is_regular}")
    print(f"  - Intervalle médian : {np.median(intervals):.2f} ms")
    print(f"  - Fréquence estimée : {sampling_rate_hz:.2f} Hz")
    print(f"  - Durée totale : {total_duration_s:.2f} secondes")
    print(f"\nParamètres suggérés:")
    print(f"  - Rééchantillonnage : {target_interval_ms:.2f} ms ({sampling_rate_hz:.2f} Hz)")
    print(f"  - Max gap : {max_gap_ms:.2f} ms")
    print(f"  - Fenêtre smoothing : {window_size} points")
    print(f"  - Durée baseline : {baseline_duration:.2f} secondes")
    
    return config

# Utilisation
auto_config = analyze_data_and_suggest_config(tobii_cleaned)
```
